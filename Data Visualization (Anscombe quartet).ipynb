```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
file_path="Dataset/anscombe.csv"
with open(file_path,"r") as file_obj:
    data=csv.DictReader(file_obj)
    x1=[]
    x2=[]
    x3=[]
    x4=[]
    y1=[]
    y2=[]
    y3=[]
    y4=[]
    for row in data:
        x1.append(row['x1'])
        x2.append(row['x2'])
        x3.append(row['x3'])
        x4.append(row['x4'])
        y1.append(row['y1'])
        y2.append(row['y2'])
        y3.append(row['y3'])
        y4.append(row['y4'])
np_x1=np.array(x1,dtype=float)
np_x2=np.array(x2,dtype=float)
np_x3=np.array(x3,dtype=float)
np_x4=np.array(x4,dtype=float)
np_y1=np.array(y1,dtype=float)
np_y2=np.array(y2,dtype=float)
np_y3=np.array(y3,dtype=float)
np_y4=np.array(y4,dtype=float)
```
