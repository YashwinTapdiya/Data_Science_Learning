### Customizing graph

```python
#Color, Marker,Linewidth
x=np.arange(0,5,0.1)
y=x**3
#plt.plot(x,y,color="black" ,marker="p")
plt.plot(x,y,color="black",linewidth=5)
plt.show()
```

```python
#(lables for axis)to get x anf y label and title
x=np.arange(0,5,0.1)
y=x**3
plt.plot(x,y)
plt.ylabel("X^3")
plt.xlabel("X")
plt.title("Matplotlib demo")
plt.show()
```

```python
#showing datasets
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color="black",label="X^3")
plt.plot(x,y2,color="red",label="X^2")
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Matplotlib demo")
plt.legend()
plt.show()
```

```python
# adding grid
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color="black",label="X^3")
plt.plot(x,y2,color="red",label="X^2")
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Matplotlib demo")
plt.legend()
plt.grid()
plt.show()
```

```python
#adding text at some point
#showing datasets
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color="black",label="X^3")
plt.plot(x,y2,color="red",label="X^2")
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Matplotlib demo")
plt.text(2,80,"text",fontsize=30)
plt.legend()
plt.grid()
plt.show()
```

```python
sizes= [3,4,6,2]
labels=["A","B","C","D"]
colors=['blue','purple','red','pink']
plt.pie(sizes , colors=colors,labels=labels,autopct="%.1f%%")
plt.axis("equal")
plt.title("Split among classes")
plt.show()
```

```python
#explode property
sizes= [3,4,6,2]
labels=["A","B","C","D"]
colors=['blue','purple','red','pink']
explode=[0.21,0,0,0]
plt.pie(sizes , colors=colors,explode=explode,labels=labels)
plt.axis("equal")
plt.title("Split among classes")
plt.show()
```

```python
# Direction
sizes= [3,4,6,2]
labels=["A","B","C","D"]
colors=['blue','purple','red','pink']
plt.pie(sizes , colors=colors,labels=labels,counterclock=False)
plt.axis("equal")
plt.title("Split among classes")
plt.show()
```

```python
#changeing start angle:
sizes= [3,4,6,2]
labels=["A","B","C","D"]
colors=['blue','purple','red','pink']
plt.pie(sizes , colors=colors,labels=labels,startangle=100)
plt.axis("equal")
plt.title("Split among classes")
plt.show()
```

### Amazon jobs dataset graphs

```python
import csv
#Plot Job_year
file_path="Dataset/amazon_jobs_dataset.csv"
with open(file_path,"r",encoding='utf-8') as file_obj:
    data=csv.DictReader(file_obj)
    d={}
    for row in data:
        s=row["Posting_date"]
        key=s[-4:]
        d[key]=d.get(key,0)+1
year=list(d.keys())
posting=list(d.values())
np_year=np.array(year,dtype=int)
np_posting=np.array(posting,dtype=int)
plt.scatter(np_year,np_posting)
plt.plot(np_year,np_posting)
plt.show()
for i in range(len(np_year)):
    print(np_year[i]," : ",np_posting[i])
```

```python
#Bar graph between Month Vs Job Opening:
file_path="Dataset/amazon_jobs_dataset.csv"
with open(file_path,"r",encoding='utf-8') as file_obj:
    data=csv.DictReader(file_obj)
    d={}
    for row in data:
        s=row["Posting_date"]
        key=s[0:3]
        d[key]=d.get(key,0)+1
month=list(d.keys())
posting=list(d.values())
np_month=np.array(month,dtype=str)
np_posting=np.array(posting,dtype=int)
plt.bar(np_month,np_posting)
plt.show()
for i in range(len(np_month)):
    print(np_month[i]," ",np_posting[i])
```

```python
#Pie-Chart-> city(india)vsjob:
file_path="Dataset/amazon_jobs_dataset.csv"
with open(file_path,"r",encoding='utf-8') as file_obj:
    data=csv.DictReader(file_obj)
    d={}
    for row in data:
        s=row['location']
        if(s[0:2]=="IN"):
            d[s]=d.get(s,0)+1
print(d)
city=list(d.keys())
posting=list(d.values())
np_city=np.array(city,dtype=str)
np_posting=np.array(posting,dtype=int)
plt.pie(np_posting,labels=np_city,autopct="%0.1f%%")
plt.show()
```
