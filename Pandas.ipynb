## Handling NaN Values

```python
df=iris.copy()
df
```

```python
#First we will add some nan values in our iris dataset using a constant from numpy library
df.iloc[2:4,1:3]=np.nan
df.head()
```

```python
df.describe()
```

```python
#dropna function is used to drop the rows which have nan entries
df.dropna(inplace =True)
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df=iris.copy()
df.columns=['sl','sw','pl','pw','flower_type']
df.iloc[2:4,1:3] = np.nan
df.head()
```

```python
#To fill the nan entries, fillna function is used
#There are several ways to fill the nan entries
```

```python
df.sw.fillna(df.sw.mean(),inplace=True)
df.head()
```

```python
df.pl.fillna(df.pl.mean(),inplace=True)
df.head()
```

### Handling Strings in Data

```python
df=iris.copy()
df.columns=['sl','sw','pl','pw','flower_type']
```

```python
df["Gender"] = "Female"
df.iloc[0:10,5]="Male"
```

```python
df.head(15)
```

```python
def f(s):
    if s=="Male":
        return 0
    else:
        return 1
df["sex"]=df.Gender.apply(f)
df.head(15)
```
