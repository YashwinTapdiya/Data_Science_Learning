## Visualizing statistical realtionships:
The process of understanding relationships between variables of a dataset and 
how these relationships, in turn, depend on other variables is known as 
statistical analysis

```python
# replot():
#This is a figure-level-function that makes use of two other axes functions for 
#Visualizing Statistical Relationships which are :
# scatterplot() and lineplot()
# by default it plots scatterplot()
```

Scatterplot

```python
#A scatterplot is the most common example of visualizing relationships between 
#two variables.

#It depicts the joint distribution of two variables using a cloud of points, where each 
#point represents an observation in the dataset.

#This depiction allows the eye to infer a substantial amount of information about 
#whether there is any meaningful relationship between them.
```

```python
df=sns.load_dataset('tips')
df.head()
#df.describe()
```

```python
sns.relplot(x='total_bill' , y = 'tip' , data=df)
plt.show()
#By default it has plotted scatter
```

```python
#now to controller what type of plot we need we can use kind(line or scatter):
sns.relplot(x='total_bill' , y = 'tip' , data=df,kind='line')
plt.show()
```

```python
sns.relplot(x='total_bill' , y = 'tip' , data=df,kind='scatter',hue='time')
plt.show()
#using hue parameter
```

```python
sns.relplot(x='total_bill' , y = 'tip' , data=df,kind='scatter',style='time')
plt.show()
#using style parameter
```

```python
sns.relplot(x='total_bill' , y = 'tip' , data=df,kind='scatter',style='time',hue='smoker')
plt.show()
#using style and hue together parameter
```

```python
sns.relplot(x='total_bill' , y = 'tip' , data=df,kind='scatter',hue='smoker',col='time')
plt.show()
#using col parameter
```

```python
#we can use scatterplot directly:
sns.scatterplot(x='total_bill' , y = 'tip' , data=df,hue='time')
plt.show()
```

# Visualizing	Categorical	Data

###### catplot()

```python
#the relationship between two variables of which one would be 
#categorical.
```

```python
# catplot() function is used to analyze the relationship between a numeric value 
#and a categorical group of values together

#Categorical data is represented in x-axis and values correspond to them 
#represented through y-axis (standard notation).
```

Categorical	Scatter	Plots

```python
#Categorical scatterplots:
# stripplot() (with kind=strip; the default)
# swarmplot() (with kind=swarm)
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
#stripplot():
sns.catplot(x='sex' ,y='total_bill' ,data=df)
plt.show()
#The first parameter is the categorical column, the second parameter is the numeric 
#column while the third parameter is the dataset.
```

```python
#jiter parameter:
#The jitter parameter controls the magnitude of jitter or disables it altogether.
sns.catplot(x='sex' ,y='total_bill' ,data=df,jitter=0.25)
plt.show()
```

```python
#swarmplot():
#Similar to stripplot, only difference is that it does not allow overlapping of 
#markers.

# adjusts the points along the categorical axis

# It can give a better representation of the distribution of observations.
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='swarm')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='swarm',hue='smoker')
plt.show()
```

Categorical Distribution plots 

```python
#when size of dataset grows, we use distribution plots 
```

```python
#Categorical distribution plots:
# boxplot() (with kind=box)
# violinplot() (with kind=violin)
# boxenplot() (with kind=boxen)
```

```python
#boxplot():
#useful when we want to compare data between two groups

#Box Plot is the visual representation of the depicting groups of numerical data 
#through their quartiles. it detects outliers
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='box')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='box',hue='smoker')
plt.show()
```

```python
#Boxen plot:
#The Boxen plot is very similar to box plot, except for the fact that it plots 
#different quartile values.
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='boxen')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='boxen',hue='smoker')
plt.show()
```

```python
#Violin plot:
#violin plot features a kernel density estimation (KDE) of the 
#underlying distribution.

#The Kernel Density Estimation is a mathematical process of finding an 
#estimate probability density function of a random variable. The estimation 
#attempts to infer characteristics of a population, based on a finite data set.
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
sns.catplot(x='species',y='sepal_length',data=df,kind='violin')
plt.show()
```

```python
sns.catplot(x='species',y='sepal_width',data=df,kind='violin')
plt.show()
```

```python
#The advantage of a violin plot is that it can show nuances in the distribution 
#that arent perceptible in a boxplot
```

### Count-Plot

```python
#way to count up the number of obs per category and then display tha info
#in bars

#useful in exploratory data analysis
```

```python
diamonds = sns.load_dataset('diamonds')

diamonds.head()
```

```python
diamonds = diamonds[diamonds.clarity.isin(['SI1', 'VS2'])]
diamonds.head()
```

```python
diamonds.shape
```

###### Basics

```python
sns.set_style('darkgrid')
```

```python
sns.countplot(x='color', data=diamonds);
```

```python
diamonds.color.value_counts(sort=False)
```

```python
sns.histplot(penguins.flipper_length_mm);
```

```python
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('axes', labelsize=14) 
```

```python
sns.set_style('white')
```

```python
sns.histplot(penguins.bill_length_mm)
sns.despine()
plt.xlabel('')
plt.tight_layout();
```

```python
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)  
```

```python
plt.figure(figsize=(7,4))
sns.histplot(penguins.bill_length_mm, alpha=1)
sns.despine(left=True)
plt.yticks([])
plt.xlabel('')
plt.ylabel('')
plt.tight_layout();
```

### Statistics

##### stat

```python
sns.histplot(x='bill_length_mm', data=penguins,
             stat='count'
            );
```

#### density - AREA of histogram sums to one

```python
sns.histplot(x='bill_length_mm', data=penguins,
             stat='density'
            );
```

###### probability - HEIGHT of histogram bars to one

```python
sns.histplot(x='bill_length_mm', data=penguins,
             stat='probability'
            );
```

#### cumulative

```python
sns.histplot(x='bill_length_mm', data=penguins,
             stat='probability', fill=False, element='step',
             cumulative=True
            );
```
